#### 2.1 Aplica una transformaci√≥n puntual. Modifica el valor de los pixeles considerando un valor de umbral (threshold). la funci√≥n de treshold debe ser una rampa con una pendiente dada (tu defines el valor de la pendiente) y cuyo valor m√≠nimo ser√° 0 y el m√°ximo ser√° 255 (valores t√≠picos de los pixeles).

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Mediante un 3 ciclos for y la funcion thereshold_transform aplicamos los tensores en la imagen original
thereshold=128
slope=1.9

def thereshold_transform(pixel_value,threshold,slope):
  if pixel_value<thereshold:
    return max(0,pixel_value*slope)
  else:
    return min(255,pixel_value*slope)



imagen_trasnformada_1=np.zeros_like(imagen)  #Creamos un arreglo matricial del mismo tama√±o de la imagen original ,pero con ceros

for i in range(imagen.shape[0]):
  for j in range(imagen.shape[1]):
    for c in range(imagen.shape[2]):
      imagen_trasnformada_1[i,j,c]=thereshold_transform(img[i,j,c],thereshold,slope)

imagen_final=np.clip(imagen_trasnformada_1,0,255).astype(np.uint8)


plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(imagen)
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(imagen_final)
plt.axis("off")

```

## 3. Realiza la inversi√≥n de imagen

### 3.1 Utilizando estrictamente numpy, busca y usa una imagen binaria y realizar la inversi√≥n de dicha imagen.
### Recuerda adjuntar la imagen a la entrega de tu trabajo.


```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def load_binary_image(imagen_binaria):
    image = Image.open(imagen_binaria).convert('L')
    image = image.point(lambda p: p > 128 and 255)
    return np.array(image)


def invert_binary_image(image):
    inverted_image = np.logical_not(image // 255).astype(np.uint8) * 255
    return inverted_image

imagen_binaria = '/content/Figura-5-Representacion-de-una-imagen-binaria.ppm'


binary_image = load_binary_image(imagen_binaria) #cargamos a imagen binaria


inverted_image = invert_binary_image(binary_image) #la invertimos con la funcion definida invert_binary_image

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Imagen Binaria Original')
plt.imshow(binary_image,cmap="gray")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Imagen Binaria Invertida')
plt.imshow(inverted_image,cmap="gray")
plt.axis('off')

plt.show()
```

## 4. Segmentaci√≥n de imagen
#### 4.1 Realiza un algoritmo para dividir una imagen con 3 canales RGB de 512x512, en parches (secciones) de 16x16. Presenta cada parche generado de la imagen.
El algoritmo debe tolerar el cambio de imagen por una de menor tamano (128, 64). Claramente, en cada caso el n√∫mero de parches ser√° menor.
    

```python


import numpy as np
import matplotlib.pyplot as plt
from PIL import Image


tamano_parche = 16

def division_imagen(imagen, tamano_parche):  #Definimos el largo y el ancho de los parches que se van a trasponer en la imagen
    largo_imagen, ancho_imagen, _ = imagen.shape
    parches = []
    for i in range(0, largo_imagen, tamano_parche):
        for j in range(0, ancho_imagen, tamano_parche):
            parche = imagen[i:i + tamano_parche, j:j + tamano_parche]
            parches.append(parche)
    return parches

def division(parches, tamano_parche): #Implementamos la division en la imagen
    numero_parches = len(parches)
    tamano_grid = int(np.ceil(np.sqrt(numero_parches)))
    fig, axes = plt.subplots(tamano_grid, tamano_grid, figsize=(10, 10))
    for ax, parche in zip(axes.flatten(), parches):
        ax.imshow(parche)
        ax.axis('off')
    plt.tight_layout()
    plt.show()


image = np.array(imagen)

patches = division_imagen(image, tamano_parche)
division(patches, tamano_parche) #llamamos nuestra funcion division la cual nos va arrojar la imagen con los parches de 16X16


```
